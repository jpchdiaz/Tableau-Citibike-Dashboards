```python
import pandas as pd
```

```python
#Q3 2016

jul_16 = pd.read_csv("data/q3_2016/201607-citibike-tripdata.csv")
aug_16 = pd.read_csv("data/q3_2016/201608-citibike-tripdata.csv")
sept_16 = pd.read_csv("data/q3_2016/201609-citibike-tripdata.csv")
```

```python
#checking
jul_16.head()
```

```python
#append and check
q3_2016 = jul_16.append([aug_16, sept_16], ignore_index=True)

q3_2016.head()
```

```python
#convert numeric values
q3_2016['gender'] = q3_2016['gender'].replace({0: 'unknown', 1: 'male', 2: 'female'})

q3_2016.head()
```
